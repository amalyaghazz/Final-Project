```python
import turtle 
#import the turtle graphics library

#draw triangles (x,y represents different points of the triangle the turtle msut mark)
def triangle(flag, x1, y1, x2, y2, x3, y3):
    flag.penup()
    flag.goto(x1, y1)
    flag.pendown()
    flag.goto(x2, y2)
    flag.goto(x3, y3)
    flag.goto(x1, y1)
    flag.penup()
    
#draws the rectangle with width and height
def rectangle(flag, x, y, width, height):
    flag.up()
    flag.goto(x, y)
    flag.setheading(0)
    flag.down()
    for i in range(2):
        flag.forward(width)
        flag.right(90)
        flag.forward(height)
        flag.right(90)
    flag.up()
    
#color the triangles
def colorTriangle(flag, x1, y1, x2, y2, x3, y3, color):
    flag.fillcolor(color)
    flag.begin_fill()
    triangle(flag,x1,y1,x2,y2,x3,y3)
    flag.end_fill()
    
turtle.setup(1500,1000,0,0)
#set up the turtle function with pixel dimensions 

#creating the colors (found on wikipedia)
colorBlue = '#003D88'
colorYellow = '#FCD955'
colorRed = '#D92323'
colorWhite = '#FFFFFF'
colorGreen = '#007A3A'


final = turtle.Turtle()
final.screen.colormode(255)
rectangle(final,0, 300,600,300)
final.goto(0,0)

#blue triangle
colorTriangle(final, 0, 0, 0,300,200,300,colorBlue)

#yellow triangle
colorTriangle(final, 0, 0, 200,300,400,300,colorYellow)

#red triangle
colorTriangle(final, 0, 0,400, 300,600,300,colorRed)

#white triangle
colorTriangle(final, 0, 0,600, 300,600,150,colorWhite)

#red triangle
colorTriangle(final, 0, 0,600, 150,600,0,colorGreen)
```
